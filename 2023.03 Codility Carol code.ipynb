```python
def get_rotation_colour(tile, left):
    if tile[0] == left:
        return 1, tile[2]
    elif tile[1] == left:
        return 2, tile[3]
    elif tile[2] == left:
        return 1, tile[0]
    else:
        return 0, tile[1]


def solution(A):  # "WBGR", "WBGR", "WRGB", "WRGB", "RBGW"
    rotations_needed = [0] * 4  # For tile 1 in 0, 90, 180, 270 deg rotations
    for ini_rot in range(4):
        latest_out = A[0][(1+ini_rot) % 4]
        rotations_needed[ini_rot] += (ini_rot if ini_rot <=2 else 1)
        for tile in A[1:]:
            p_rot, latest_out = get_rotation_colour(tile, latest_out)
            rotations_needed[ini_rot] += p_rot

    return min(rotations_needed)
```
