# Codility - Year of the rabbit challenge

```python
def is_OK(A, B):
    return all([a != b for (a, b) in zip(A, B)])

def solution(A, B):
    B_rotated = B
    for rot in range(len(A)):
        if is_OK(A, B_rotated):
            return rot
        B_rotated = B_rotated[-1:] + B_rotated[:-1]
    return -1
```

```python
solution([1, 3, 5, 2, 8, 7], [7, 1, 9, 8, 5, 7])
```

```python
solution([1, 3, 5, 2, 8, 7], [7, 7, 1, 9, 8, 5])
```
